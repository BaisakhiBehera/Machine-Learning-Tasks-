```python
# 	Identify numerical columns  and categorical_columns in the dataset.
numerical_columns = df.select_dtypes(include=["int64","float64"])
categorical_columns = df.select_dtypes(include=["object"])
print(numerical_columns)
print(categorical_columns)
```

```python
#Plot histograms for each numerical column.
matplotlib.pyplot as plt
for col in numerical_columns:
    plt.hist(df[col])
    plt.title("Histogram of " + col)
    plt.show()
```

```python
# Analyze skewness
# Calculate skewness for numerical columns 
numerical_columns.skew()
```

```python
"""Boxplots:
	Create boxplots for numerical variables to identify outliers and the interquartile range.
	Discuss any findings, such as extreme values or unusual distributions"""
for col in numerical_columns:
    plt.figure()
    df.boxplot(column=col, vert=False)
    plt.title(f"Boxplot of '{col}'")
    plt.xlabel(col)
    plt.show()
```

```python
for col in numerical_columns:
    IQR = df[col].describe()['75%'] - df[col].describe()['25%']
    print(IQR)
    LW = df[col].describe()['25%'] - (1.5 * IQR)
    UW = df[col].describe()['75%'] + (1.5 * IQR)
    outliers =  df[col][(df[col] < LW) | (df[col] > UW)]
    print(outliers)
    print(len(outliers))
```

```python
#â—	Bar Chart Analysis for Categorical Column:
for i in categorical_columns:
    plt.figure()
    df[i].value_counts().plot(kind="bar")
    plt.title(f"Distribution of '{i}'")
    plt.xlabel(i)
    plt.ylabel("Count")
    plt.show()
```

```python
# Explain the concept of standardization (z-score normalization).

# Standardization is a technique used to transform numerical data to have a mean of 0 and a standard deviation of 1.
# This process is also known as z-score normalization.
# Standardization is useful for comparing data points that have different units or scales.
numerical_columns = df.select_dtypes(include=["int64", "float64"]).columns.tolist()
df_standardized = df.copy()
for col in numerical_columns:
  df_standardized[col] = (df_standardized[col] - df_standardized[col].mean()) / df_standardized[col].std()

# Show before and after comparisons of the data distributions

# Before standardization
print("Before standardization:")
df[numerical_columns].describe()


```

```python

# After standardization
print("After standardization:")
df_standardized[numerical_columns].describe()
```

```python
#	Apply one-hot encoding to the categorical columns, creating binary (0 or 1) columns for each category.

# Select categorical columns
categorical_columns = df.select_dtypes(include=["object"]).columns.tolist()

# One-hot encode categorical columns
df_encoded = pd.get_dummies(df, columns=categorical_columns)

# Display the entire transformed dataset
print(df_encoded)
```
